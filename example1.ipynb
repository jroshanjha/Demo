```python
def tower_of_hanoi(n, source, target, auxiliary):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    tower_of_hanoi(n - 1, source, auxiliary, target)
    print(f"Move disk {n} from {source} to {target}")
    tower_of_hanoi(n - 1, auxiliary, target, source)

n = 4 
tower_of_hanoi(n, 'A', 'C', 'B')
```

```python
import statistics as sts 
import scipy.stats as stats
import numpy as np 
mean = np.mean(data) # 103.05 


sts.stdev(data)  # 8.537687348641779
# The population standard deviation is the square root of the sample standard deviation.
population_standard_deviation = np.sqrt(sts.stdev(data)**2/(len(data)-1)) # 9.022100705173285   

# Confidence level: A confidence level of 95% means that we are 95% confident that our sample mean is within the range of 99.5% of the population mean.
confidence_level = 0.95
margin_of_error = population_standard_deviation * np.sqrt(1/len(data) + (1/(2*len(data))) * (1-1/len(data)))
confidence_interval = margin_of_error * stats.t.ppf((1+confidence_level)/2, len(data)-1)
print(f'Confidence interval: {confidence_interval:.2f}')
```

```python
# K Means Clustering Algorithms Implementation .....
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
%matplotlib inline 
from sklearn.datasets import make_blobs

# make blobs ->  n_sample= 100,n_feature = 2, Center = None ) 
# Generate isotropic Gaussian blobs for clustering
```

```python
X,y = make_blobs(n_samples=1000, n_features=2, centers=3,random_state=23)
X.shape
```

```python
# plt.scatter(X[:,0],X[:,1],c=y)
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.model_selection import train_test_split 

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.cluster import KMeans 

# Manule Processs 

## Elbow Method  to select the K value 
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
## Plot for the Elbow curve 
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

```

```python
kmeans =  KMeans(n_clusters=3,init='k-means++')
y_labes = kmeans.fit_predict(x_train)
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=y_labes)
```

```python
y_labes1 = kmeans.predict(x_test)
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c=y_labes1)
```

```python
## knee Locator 
# pip install kneed
from kneed import KneeLocator 

kl = KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
kl.elbow
```

```python
##  Performance metrics 
##  Silhoutte score  for huge amount of data.....

from sklearn.metrics import silhouette_score
silhouette_score_val =[]
for k in range(2,11):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train)
    score = silhouette_score(x_train,kmeans.labels_)
    silhouette_score_val.append(score)
    
```

```python
silhouette_score_val
```

```python
## Plot for the Elbow curve 
plt.plot(range(2,11),silhouette_score_val)
plt.xticks(range(2,11))
plt.xlabel('Number of Clusters')
plt.ylabel('Silhoutte Coeffecient')
plt.show()

```

```python
from scipy import stats
import numpy as np

# Sample data
sample_data = [5.1, 4.9, 4.7, 4.6, 5.0, 5.4, 4.6, 5.0, 4.4, 4.9]

# Hypothesized population mean
mu_0 = 5.0

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(sample_data, mu_0)

# Print the results
print(f"t-statistic: {t_statistic:.4f}")
print(f"p-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis (sample mean is significantly different from the hypothesized mean)")
else:
    print("Fail to reject the null hypothesis (sample mean is not significantly different from the hypothesized mean)")
```
